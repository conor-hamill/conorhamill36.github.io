```python
import folium
```

```python
min_lat = -60.0
```

```python
locations_dict = {
    "Tokyo": [35.6764, 139.6500],
    "Lisbon": [38.7223, -9.1393],
    "Charlotte": [35.2271, -80.8431]
}
```

```python
m = folium.Map(
    max_bounds=True,
    min_lat=min_lat,
    min_zoom=1,
    max_zoom=16
)

for key, value in locations_dict.items():
    lat, lon = value
    folium.Marker([lat, lon], popup=key).add_to(m)
```
